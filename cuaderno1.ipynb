```python
from faker import Faker
# Crear instancia de Faker
fake = Faker()
```

```python
lista=[]
```

```python
for i in range(0,10000):
    lista.append({
        'nombre': fake.name(),
        'apellido': fake.last_name(),
        'email': fake.email(),
        'direccion': fake.address()
    })
```

```python
lista
```

```python
df = pd.DataFrame(lista)
```

```python
import random
```

```python
# Provincias permitidas
provincias = ['Loja', 'Pichincha', 'Guayas', 'Azuay', 'Manab']

# Agregar edad y provincia
df['edad'] = [fake.random_int(min=30, max=40) for j in range(len(df))]
df['provincia'] = [random.choice(provincias) for x in range(len(df))]
```

```python
df
```

```python
df.to_csv('conjunto_datos.csv')
```
